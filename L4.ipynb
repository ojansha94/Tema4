```python
# Solucin del Laboratorio 4

# Los parmetros T, t_final y N son elegidos arbitrariamente

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
#phi = varianza 
phi= 8
w0= np.pi
# Variables aleatorias x y Y
vaX = stats.norm(0, np.sqrt(phi))
vaY = stats.norm(0, np.sqrt(phi))
# Creacin del vector de tiempo
T = 100			# nmero de elementos
t_final = 10	# tiempo en segundos
t = np.linspace(0, t_final, T)

# Inicializacin del proceso aleatorio X(t) con N realizaciones
N = 10
X_t = np.empty((N, len(t)))	# N funciones del tiempo x(t) con T puntos

# Creacin de las muestras del proceso x(t) (A y Z independientes)
for i in range(N):
	X = vaX.rvs()
	Y = vaY.rvs()
	w_t = X * np.cos(w0*t) + Y * np.sin(w0*t)
	X_t[i,:] = w_t
	plt.plot(t, w_t)

# Promedio de las N realizaciones en cada instante (cada punto en t)
P = [np.mean(X_t[:,i]) for i in range(len(t))]
plt.plot(t, P, lw=6)

# Graficar el resultado terico del valor esperado
E = 0*t 
plt.plot(t, E, '-.', lw=4)

# Mostrar las realizaciones, y su promedio calculado y terico
plt.title('Realizaciones del proceso aleatorio $X(t)$')
plt.xlabel('$t$')
plt.ylabel('$x_i(t)$')
plt.show()

# T valores de desplazamiento tau
desplazamiento = np.arange(T)
taus = desplazamiento/t_final

# Inicializacin de matriz de valores de correlacin para las N funciones
corr = np.empty((N, len(desplazamiento)))

# Nueva figura para la autocorrelacin
plt.figure()

# Clculo de correlacin para cada valor de tau
for n in range(N):
	for i, tau in enumerate(desplazamiento):
		corr[n, i] = np.correlate(X_t[n,:], np.roll(X_t[n,:], tau))/T
	plt.plot(taus, corr[n,:])

# Valor terico de correlacin
Rww = phi*np.cos(w0*taus)

# Grficas de correlacin para cada realizacin y la
plt.plot(taus, Rww, '-.', lw=4, label='Correlacin terica')
plt.title('Funciones de autocorrelacin de las realizaciones del proceso')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$R_{WW}(\tau)$')
plt.legend()
plt.show()

# Grfica del valor medio
plt.plot(taus, E, '-.', lw=4, label='Valor terico de la media')
plt.title('Valor terico de la media')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$E[W(\tau)])(\tau)$')
plt.legend()
plt.show()
```
